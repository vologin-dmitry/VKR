```python
# Данные по загрязнениям
df_2 = pd.read_csv('/content/drive/MyDrive/data_environment_113_v20230123.csv', sep = ';', decimal='.', thousands=',')
df_2.rename({'object_name': 'region'}, axis='columns', inplace = True)
df_2 = df_2[['indicator_name', 'region', 'year', 'indicator_value']]
df_2
```

```python
x_1 = df_2[(df_2['indicator_name'] == 'Население в городах с высоким и очень высоким уровнем загрязнения воздуха, % городского населения')]
x_2 = df_2[(df_2['indicator_name'] == 'Выбросы от стационарных источников (на душу населения)')]
x_3 = df_2[(df_2['indicator_name'] == 'Выбросы от передвижных источников (на душу населения)')]
x_4 = df_2[(df_2['indicator_name'] == 'Доля водопроводов хозяйственно-питьевого централизованного водоснабжения, не соответствующих санитарным нормам и правилам')]
x_5 = df_2[(df_2['indicator_name'] == 'Сброс загрязненных сточных вод (на душу населения)')]
x_6 = df_2[(df_2['indicator_name'] == 'Расходы на охрану окружающей среды')]
x_7 = df_2[(df_2['indicator_name'] == 'Количество особо охраняемых природных территорий (всего)')]
x_8 = df_2[(df_2['indicator_name'] == 'Площадь особо охраняемых природных территорий (всего)')]
x_9 = df_2[(df_2['indicator_name'] == 'Образовано отходов производства и потребления')]
x_10 = df_2[(df_2['indicator_name'] == 'Образовано твердых коммунальных отходов (всего)')]
x_11 = df_2[(df_2['indicator_name'] == 'Захоронено твердых коммунальных отходов (на душу населения)')]
```

```python
X = x_1[['region', 'year', 'indicator_value']].rename(columns={'indicator_value': 'x_1'})

X = X.merge(x_2[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_2'})

X = X.merge(x_3[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_3'})

X = X.merge(x_4[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_4'})

X = X.merge(x_5[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_5'})

X = X.merge(x_6[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_6'})

X = X.merge(x_7[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_7'})

X = X.merge(x_8[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_8'})

X = X.merge(x_9[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_9'})

X = X.merge(x_10[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_10'})

X = X.merge(x_11[['region', 'year', 'indicator_value']], on=['region', 'year'], how='outer')
X = X.rename(columns={'indicator_value': 'x_11'})
X.drop_duplicates(inplace=True)
X
```

```python
# Данные по смертности
df_3 = pd.read_csv('/content/drive/MyDrive/data_death_rates_116_v20240327.csv', sep = ';', decimal='.', thousands=',')
df_3 = df_3[['region', 'year', 'sex', 'age_group', 'subclass', 'death_count_per100k']]
```

```python
y_1 = df_3[(df_3['subclass'] == 'Болезни щитовидной железы') & (df_3['age_group'] == 60) & (df_3['sex'] == 'M')]
y_2 = df_3[(df_3['subclass'] == 'Сахарный диабет') & (df_3['age_group'] == 60) & (df_3['sex'] == 'M')]
y_3 = df_3[(df_3['subclass'] == 'Органические, включая симптоматические, психические расстройства') & (df_3['age_group'] == 60) & (df_3['sex'] == 'M')]
y_4 = df_3[(df_3['subclass'] == 'Расстройства настроения [аффективные расстройства]') & (df_3['age_group'] == 60) & (df_3['sex'] == 'M')]

Y = y_1[['region', 'year', 'death_count_per100k']].rename(columns={'death_count_per100k': 'y_1'})
Y = Y.merge(y_2[['region', 'year', 'death_count_per100k']], on=['region', 'year'], how='outer')
Y = Y.rename(columns={'death_count_per100k': 'y_2'})
Y = Y.merge(y_3[['region', 'year', 'death_count_per100k']], on=['region', 'year'], how='outer')
Y = Y.rename(columns={'death_count_per100k': 'y_3'})
Y = Y.merge(y_4[['region', 'year', 'death_count_per100k']], on=['region', 'year'], how='outer')
Y = Y.rename(columns={'death_count_per100k': 'y_4'})
Y.drop_duplicates(inplace=True)
Y
```

```python
df_model = X.merge(Y, on = ['region', 'year'], how = 'left')
df_model = df_model[df_model['region'] != 'Российская Федерация']
numeric_cols = df_model.select_dtypes(include=['int64', 'float64']).columns
df_model[numeric_cols] = df_model[numeric_cols].fillna(df_model[numeric_cols].mean())
object_cols = df_model.select_dtypes(include=['object']).columns
df_model[object_cols] = df_model[object_cols].fillna('')
df_model
```

```python
import pandas as pd
import numpy as np

# предположим, что данные уже загружены в dataframe df_model
regions = df_model['region'].unique()
np.random.seed(42) # фиксируем сид для воспроизводимости
random_regions = np.random.choice(regions, size=5, replace=False)
```

```python
df_selected = df_model[df_model['region'].isin(random_regions)]

X = df_selected.drop(['region', 'year', 'y_1', 'y_2', 'y_3', 'y_4'], axis=1)
y = df_selected[['y_1', 'y_2', 'y_3', 'y_4']]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor

regions = df_model['region'].unique()
np.random.seed(42)
selected_regions = np.random.choice(regions, size=5, replace=False)

for region in selected_regions:
    print(f'Регион: {region}')
    df_region = df_model[df_model['region'] == region]

    X = df_region.drop(['region', 'year', 'y_1', 'y_2', 'y_3', 'y_4'], axis=1)
    y = df_region[['y_1', 'y_2', 'y_3', 'y_4']]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    for i in range(1, 5):
        model = GradientBoostingRegressor(random_state=42)
        model.fit(X_train, y_train.iloc[:, i-1])

        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test.iloc[:, i-1], y_pred)
        r2 = r2_score(y_test.iloc[:, i-1], y_pred)
        feature_importance = model.feature_importances_

        print(f'Модель для y_{i}')
        print(f'MSE: {mse.round(2)}')
        print(f'R2: {r2.round(2)}')
        print(f'Важность признаков: {feature_importance.round(2)}')
        print()
```
